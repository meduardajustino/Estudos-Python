```python
import numpy as np
dados_PE = np.genfromtxt('temperatura_PE.csv', delimiter=',', skip_header=1)

dados_PE = np.transpose(dados_PE)
dados_PE
```

```python
temperatura_pe = dados_PE[1]
temperatura_pe
```

```python
print('Como foi solicitado, seguem os dados estatísticos descritivos a respeito da análise das temperaturas por hora do meu país Pernambuco <3: ')

#MÉDIA
media_temp = np.mean(temperatura_pe)
print(f'A média das temperaturas é de {media_temp:.3f}')

#MEDIANA
mediana_temp = np.median(temperatura_pe)
print(f'A mediana das temperaturas é de {mediana_temp:.3f}')

#DESVIO PADRÃO
dp_temp = np.std(temperatura_pe)
print(f'O desvio padrão das temperaturas é de {dp_temp:.3f}')

#QUARTIS
MIN = np.min(temperatura_pe)
Q1 = np.quantile(temperatura_pe,0.25)
Q2 = np.quantile(temperatura_pe,0.5)
Q3 = np.quantile(temperatura_pe,0.75)
MAX = np.max(temperatura_pe)

print(f'Valor mínimo:" {MIN:.3f}')
print(f'Primeiro Quartil(25%):{Q1:.3f}')
print(f'Mediana/Segundo Quartil (50%): {Q2:.3f}')
print(f'Terceiro Quartil(75%):{Q3:.3f}')
print(f'Valor máximo:{MAX:.3f}')
```

```python
#BOXPLOT
import matplotlib.pyplot as plt

plt.ylabel('Temperatura em °C')
plt.xlabel('Pernambuco')
plt.title('Temperatura de Pernambuco por hora')

bp = plt.boxplot(temperatura_pe, patch_artist=True)

# Change the color of the boxes
for box in bp['boxes']:
    box.set(facecolor='#ffcbdb')

plt.show()
```

```python
plt.hist(temperatura_pe, bins=10, color='pink')
plt.ylabel('Frequência')
plt.xlabel('Temperatura em °C')
plt.title('Distribuição de Temperatura em Pernambuco')
plt.show()
```
