## Ler e extrair informações de pdf com a biblioteca PyPDF2

```python
# Lista com arquivos dentro de uma pasta
import os

caminho = r"M:\Maria Eduarda\Downloads\3R Petroleum Oleo e Gas SA"

# lista os arquivos no diretório
arquivos = os.listdir(caminho)

for arquivo in arquivos:
    print(arquivo)
```

```python
import PyPDF2

# definir caminho do arquivo pdf
local_arquivo = r"M:\Maria Eduarda\Downloads\3R Petroleum Oleo e Gas SA\2019.pdf") #sempre que tiver contra barra é importante usar o r'', pois desconsidera a contra barra como caractere de escape
# abrir e ler o arquivo pdf
if os.path.exists(local_arquivo):
    with open(local_arquivo,'rb') as arquivo: # rb para o modo de leitura binária do arquivo pdf
        leitor_pdf = PyPDF2.PdfReader(arquivo)
        texto = "" # sem valor porque é inserido durante a extração
        for pagina in leitor_pdf.pages:
            texto += pagina.extract_text()
            
        # exibição do conteúdo do arquivo pdf
        print("Conteúdo do arquivo PDF")
        print(texto)
else:
    print(f'O arquivo {local_arquivo} não foi encontrado.')
        
```

```python
print(f'Número de páginas: {len(leitor_pdf.pages)}')
```

```python
# leitura página 1
page1 = leitor_pdf.pages[0]
page1_text = page1.extract_text()

print(page1_text)
```

```python
# para retirar as quebras de linha
import re

page1_text = re.sub('\n', '', page1_text)
print(page1_text)
```

### Merge entre dois arquivos pdf

```python
arq1 = open(r"M:\Maria Eduarda\Downloads\3R Petroleum Oleo e Gas SA\2019.pdf", 'rb')
arq2 = open(r"M:\Maria Eduarda\Downloads\3R Petroleum Oleo e Gas SA\2020.pdf", 'rb')

dados_arq1 = PyPDF2.PdfReader(arq1)
dados_arq2 = PyPDF2.PdfReader(arq2)

# declarando um objeto tipo merge
merge = PyPDF2.PdfMerger()
# dando appende no merge dos meus dados de pdf
merge.append(dados_arq1)
merge.append(dados_arq2)
# escrevendo um novo arquivo pdf
merge.write(r"M:\Maria Eduarda\Downloads\3R Petroleum Oleo e Gas SA\merge.pdf")
```
