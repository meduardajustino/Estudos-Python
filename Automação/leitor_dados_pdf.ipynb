Leitor de dados de arquivos em pdf

```python
import PyPDF2
from PyPDF2 import PdfReader
import re
```

```python
file = open(r"M:\Maria Eduarda\Downloads\Hapvida\2021.pdf", 'rb')
pdf = PdfReader(file)
num_pages = len(pdf.pages)
pages = pdf.pages
```

```python
pages[0].extract_text()
```

```python

# Ler e converter o texto de cada página para minúsculas
for page_num in range(num_pages):
    page = pdf.pages[page_num]
    text = page.extract_text()
    text_lower = text.lower()
    print(text_lower)
```

```python
# Verificando se há informações de covenants
for page in pages:
    content = page.extract_text()
    if 'covenants' or 'covenant' or "cláusulas contratuais" or "cláusulas restritivas" in content:
        lines = content.split('\n')
        
        for line in lines:
            if 'covenants' in line:
                dados = line.split()
                print(line)
                print(dados)
        # print(content) adicionar em uma pasta com caminho do nome da empresa e ano
```

```python
# utilizando regex para encontrar valores como receita líquida
regex1 = 'covenants'

for page in pages:

    content = page.extract_text()
    
    match1 = re.search(regex1, content)
    if match1:
        print(match1[0])
        print(content)
```
