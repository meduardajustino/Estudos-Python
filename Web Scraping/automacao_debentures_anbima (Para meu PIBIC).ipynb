```python
# função para verificar se há "Sumário definitivo" clicar no xpath do botão "baixar"
def buscar_e_clicar_sumario(navegador):
    try:
        # Espera o elemento com o texto "Documentos da oferta" e "Sumário definitivo" estar presente na página
        WebDriverWait(navegador, 7).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), 'Documentos da oferta')]")))
        WebDriverWait(navegador, 6).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), 'Sumário definitivo')]")))

        # Localiza o botão de "baixar"
        baixar_botao = WebDriverWait(navegador, 8).until(
            EC.element_to_be_clickable( (By.XPATH, "//*[contains(text(), 'Sumário definitivo')]/following::a[contains(text(), 'Baixar')][1]"))
        )

       # ter certeza que o botão foi clicado com selenium
        baixar_botao.click()
        print("Botão 'baixar' da Anbima clicado.")
        
        # redirecionado para a página do S3 (Amazon) e aguarda 5 segundos para o download começar
        time.sleep(3)
        pyautogui.click(x=3731, y=166)  # salvar documento
        time.sleep(2) #espera 3 segundos antes de pressionar "Enter"
        pyautogui.press('enter')
        print("Download iniciado.")
        time.sleep(2) 
        pyautogui.hotkey('ctrl', 'w')
        print("Aba do S3 fechada.")

        return True  # sumário foi encontrado e o download foi iniciado
    
    except Exception as e:
        print(f"Erro ao tentar encontrar o 'Sumário definitivo' ou clicar no botão: {e}")
        navegador.back()
        return False   # não encontrou o sumário ou não conseguiu clicar no botão, ir para próxima empresa
```

```python
#  Ter noção dos debentures disponíveis pelo contador para empresas com "Sumário definitivo" de todas empresas coletadas
empresas_com_sumarios = 0
```

```python
# Iterar pelas empresas
for i, empresa in enumerate(empresas_page4_5):
    try:
        # abrindo link da empresa listada
        link_empresa = f"https://data.anbima.com.br/debentures/{empresa}/documentos"
        navegador.get(link_empresa)
        print(f"Abrindo o link da empresa {empresa} ({i + 1}/{len(empresas_page4_5)})")
        
        # buscando o "Sumário definitivo" com a def criada anteriormente e clicar no botão "baixar"
        if buscar_e_clicar_sumario(navegador):
            empresas_com_sumario += 1
        else:
            print(f"Sumário definitivo não encontrado para a empresa {empresa}")

    except Exception as e:
        print(f"Erro ao processar a empresa {empresa}: {e}")
    
    finally:
        navegador.back() 
        time.sleep(4) 

print(f'\nTotal de empresas com "Sumário definitivo" encontrado: {empresas_com_sumario}')
```
