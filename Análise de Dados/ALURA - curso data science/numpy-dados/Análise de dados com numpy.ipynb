### Preços das maçãs na Rússia

```python
import numpy as np

dado = np.loadtxt(r'M:\{estudos}\Python\Estudos-Python\Análise de Dados\ALURA - curso data science\numpy-dados\apples_ts.csv', delimiter = ",", usecols=np.arange(1,88,1))
dado
```

```python
dado_transposto = dado.T
datas = dado_transposto[:,0]
precos = dado_transposto[:,1:6]
precos
```

```python
import matplotlib.pyplot as plt
```

```python
datas = np.arange(1,88,1) # datas em meses
```

```python
plt.plot(datas, precos[:,0])
```

```python
# Selecionando as cidades e suas respectivas colunas
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
```

```python
# Separar meses em anos para analisar qual mês correspondente, array unidimensional
Moscow_ano1 = Moscow[0:12]
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]
plt.plot(np.arange(1,13,1), Moscow_ano1)
plt.plot(np.arange(1,13,1), Moscow_ano2)
plt.plot(np.arange(1,13,1), Moscow_ano3)
plt.plot(np.arange(1,13,1), Moscow_ano4)
plt.title('Preço das maçãs')
plt.legend(['ano 1', 'ano 2', 'ano 3', 'ano 4'])
```

```python
# Comparando o preço médio de Moscow e Petersburg
media_Moscow = np.mean(Moscow)
media_Petersburg = np.mean(Petersburg)
print(f'A comparação das médias dos preços das maçãs é que em Moscow ocorre {media_Moscow:.2f} e em Petersburg é de {media_Petersburg:.2f}.')
```

```python
# Inferir a taxa de crescimento do preço com regressão linear
x = datas # em meses
y = 0.52*x + 80

np.sqrt(np.sum((np.power(Moscow-y, 2))))
#raiz quadrada da soma do quadrado dos resíduos ou usar a função np.linalg.norm(Moscow-y)
plt.plot(datas, Moscow)
plt.plot(x, y)
```

```python
print(np.sqrt(np.sum((np.power(Moscow-y, 2)))))
#raiz quadrada da soma do quadrado dos resíduos ou usar a função
np.linalg.norm(Moscow-y)
```

```python
# Calculando os coeficientes da reta (regressão linear simples)
Y = Moscow
X = datas
n = np.size(Moscow)
coef_angular = (n*np.sum(X*Y)-np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
coef_angular
```

```python
# coeficiente linear/vertical
b = np.mean(Y) - coef_angular*np.mean(X)
y = coef_angular*X+b
```

```python
np.linalg.norm(Moscow-y) # Melhor ajuste!! 101 melhor que o 110
```

```python
plt.plot(datas, Moscow)
plt.plot(x,y)
plt.plot(100,100*coef_angular+b, '*r') #no mês 100, prevendo
```

### Números aleatórios para inferir regressão linear simples

```python
coef_angulares = np.random.uniform(low=0.1, high=0.9, size=100)
# Calcular a norma
norma = []
for i in range(100):
    norma = np.append(norma, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
norma
```

```python
# Reprodutibilidade com sementes
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.1, high=0.9, size=100)
# Calcular a norma
norma = []
for i in range(100):
    norma = np.append(norma, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
norma
```

```python
# Reprodutibilidade com sementes
np.random.seed(16) #funciona apenas na mesma célula usada
np.random.uniform(low=0.1,high=0.9,size=100)
dados = np.column_stack([norma, coef_angulares]) #pega arrays unidimensionais e agrega em array bidimensional

np.savetxt('dados.csv',dados,delimiter=',')
dados.shape #bidimensional
```
