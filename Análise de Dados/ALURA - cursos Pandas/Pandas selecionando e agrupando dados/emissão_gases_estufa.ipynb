### Modificando o formato do DataFrame

```python
colunas_info = list(emissoes_gases.loc[:, 'Nível 1 - Setor':'Produto'].columns)
colunas_info
```

```python
colunas_anos = list(emissoes_gases.loc[:,1970:2021].columns)
```

```python
# mudança do formato wide para o formato long com o método melt()
emissoes_ano = emissoes_gases.melt(id_vars= colunas_info, value_vars= colunas_anos, var_name= 'Ano', value_name='Emissão')
emissoes_ano
```
