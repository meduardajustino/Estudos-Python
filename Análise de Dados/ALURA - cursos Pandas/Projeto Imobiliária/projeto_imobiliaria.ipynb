## Tratando e filtrando dados

### Lidando com dados nulos

```python
df.isnull().sum()
```

```python
df = df.fillna(0) #substituindo os valores nulos por 0
```

```python
#Remover registros inconsistentes
registros_a_remover = df.query('Valor == 0' or 'Condominio == 0').index
df.query('Valor == 0' or 'Condominio == 0') # precisamos remover essas linhas
```

```python
df.drop(registros_a_remover, axis=0, inplace=True) #axis=0 remove linhas e =1 remove colunas
df.query('Valor == 0' or 'Condominio == 0')
```

```python
df.drop('Tipo', axis=1, inplace=True)
df.head()
```

### Aplicando filtros
- apartamentos com 1 quarto e aluguel menor que R$1200;
- apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$3000 e área maior que 70m².

```python
# apartamentos com 1 quarto e aluguel menor que R$1200
selecao1 = df['Quartos'] == 1
selecao2 = df['Valor'] < 1200
selecao_final = (selecao1) & (selecao2)
df_1 = df[selecao_final]
df_1.head()
```

```python
# apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$3000 e área maior que 70m²
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df_2 = df[selecao]
df_2.tail()
```

```python
# Salvar os dados em csv
df.to_csv('dados_apartamento.csv', index=False, sep=";")
```

## Manipulando os dados
- Atendendo as demandas do time dev

```python
dados = pd.read_csv(url, sep=';')
dados.tail()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados['Valor_por_ano'] = dados['Valor_por_mes']*12 + dados['IPTU']
dados.tail()
```

```python
# Criar colunas categóricas

dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quarto(s) ' + \
                     ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'

dados['Descrição'].head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```
